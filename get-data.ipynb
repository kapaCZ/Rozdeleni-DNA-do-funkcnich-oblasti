```python
import fastaparser


def Change_base(i):
    switcher = {
        "A": "T",
        "T": "A",
        "C": "G",
        "G": "C",
        "a": "t",
        "t": "a",
        "c": "g",
        "g": "c",
        }   
    return switcher.get(i,i)



def rotate_seq(dataToRotate: str):
    reversedstring = dataToRotate[::-1]
    new_string = []
    for base in reversedstring :
        new_base = Change_base(base)
        new_string.append(new_base)
    new_data = "".join(new_string)
    return new_data


import os
dir_list = os.listdir("E:/VYSOKA/BAKA/data")
dir_list2 = os.listdir("E:/VYSOKA/BAKA/data/Candida_albicans/")

def find_dirs():
    dir_list = os.listdir("E:/VYSOKA/BAKA/data")
    for dir in dir_list:
        dir_list2 = os.listdir("E:/VYSOKA/BAKA/data/" +dir)
        for file in dir_list2:
            if "ctss" not in file.split('.')[1] : continue
            if "2" in file : continue
            for file2 in dir_list2:
                if "fna" != file2.split(".")[-1] : continue
                fnafile = file2
            yield(["E:/VYSOKA/BAKA/data/" +dir +"/" + file,"E:/VYSOKA/BAKA/data/" +dir +"/" + fnafile])
    

def tss_data(d):
    tssdata = {}
    seqence = {}
    for file in d:
        print(file)
        path_to_fna = file[1]
        path_to_tss = file[0]

        with open(path_to_tss,"r") as f:
            lines_tss = f.readlines()
        
        for line in lines_tss:
            l = line.split("\t")
            if l[0] not in tssdata:
                tssdata[l[0]] = [[l[1],l[2]]]
                continue
        
            tssdata[l[0]].append([l[1],l[2]])

        with open(path_to_fna, "r") as rf:
            parser = fastaparser.Reader(rf, parse_method="quick")
            for data in parser:
                dataheader = data.header.split(" ")[0].replace('>', '')
                if dataheader not in tssdata : continue
                if dataheader not in seqence:
                    seqence[dataheader] = []
                
                emptysten = []
                for tss in tssdata[dataheader]:
                    if tss[1] == "-":
                        string = data.sequence[int(tss[0])-700:int(tss[0])+300]
                        if len(string) == 0 : emptysten.append(tss)
                        string = rotate_seq(string)
                        string = [string,tss[1]]
                        seqence[dataheader].append(string)
                        continue
                    
                    string = [data.sequence[int(tss[0])-299:int(tss[0])+701],tss[1]]    
                    seqence[dataheader].append(string)


        newFilePath = file[0].split("/")
        newFilePath[-1] = newFilePath[-2] +"_promotor_tss.txt"
        newFilePath = ("/").join(newFilePath)
        with open(newFilePath, "w") as f :
            for key in seqence.keys():
                for data in seqence[key]:
                    f.write("\n" +key+ '   ' + data[1])
                    line = data[0]
                    n = 80
                    newline = [line[i:i+n] for i in range(0, len(line), n)]
                    for line in newline:
                        f.write("\n" + line)
                    
            
            


        

                

        
        
        
d=find_dirs()          
tss_data(d)
```
